```python
import pandas as pd
```

```python
df = pd.read_csv("stops.txt")
```

```python
df = df.drop(df.index[4685:])
```

```python
df['stop_id'] = df['stop_id'].str[-4:]
df.head()
```

```python
# Casting stop_id to integers in order to remove the prefixed 0 values
# and to be the same 
df['stop_id'] = df['stop_id'].astype('int64')
```

```python
df['stop_lat'] = round(df['stop_lat'], 5)
df['stop_lon'] = round(df['stop_lon'], 5)
```

```python
df2 = pd.read_csv('static_fix.csv', header=None)
```

```python
cols = ['stopid','name','lat','long']
df3 = pd.DataFrame(columns=cols, index=df.index)
df3
```

```python
j=0
for i in range(len(df)):
    if df['stop_id'][i] not in df2[3].unique():
        df3['stopid'][j] = df['stop_id'][i]
        df3['name'][j] = df['stop_name'][i]
        df3['lat'][j] = df['stop_lat'][i]
        df3['long'][j] = df['stop_lon'][i]

    j+=1
        
    
```

```python
df3 = df3.dropna()
```

```python
df3.index = range(len(df3))
```

```python
for i in range(len(df3)):
    if df3['shortname'][i] == None:
        df3['shortname'][i] = df3['longname'][i]
    elif df3['longname'][i] == None:
        df3['longname'][i] = df3['shortname'][i]
```

```python
df3 = df3.drop(['name'], axis=1)
```

```python
df3.to_csv('missing_stop_data.csv', header=None, index=False)
```
